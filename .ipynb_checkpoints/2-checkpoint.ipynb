```python
from sklearn.datasets import load_iris
import pandas as pd
import numpy as np
```

```python
data=load_iris()
```

```python
data
```

```python
data.data
```

```python
data1=pd.DataFrame(data.data,columns=data.feature_names)
```

```python
data1
```

```python
data1['species']=data.target
```

```python
data1
```

```python
for i in data1['species']:
    if i==0:
        data1.species.replace(0,"setosa",inplace=True)
    elif i==1:
        data1.species.replace(1,"versicolor",inplace=True)
    elif i==2:
        data1.species.replace(2,"virginica",inplace=True)
    else:
        print("waste")
```
