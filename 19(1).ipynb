```python
for i in df1['Education Level']:
    if i=="Master's Degree" :
        df1["Education Level"].replace(i,"Master's",inplace=True)
    elif i=="Bachelor's Degree" :
        df1["Education Level"].replace(i,"Bachelor's",inplace=True)
    elif i=="phD" :
        df1["Education Level"].replace(i,"PhD",inplace=True)
    
```
